```python
class SchoolManagement:
    def __init__(self):
        self.students = {}
        self.next_id = 1

    def validate_mobile(self, mobile):
        return mobile.isdigit() and len(mobile) == 10


    def validate_age(self, age):
        return 5 <= age <= 18


    def new_admission(self):
        name = input("Enter student name: ")
        age = int(input("Enter age: "))
        if not self.validate_age(age):
            print("Invalid age! Must be between 5 and 18.")
            return

        student_class = input("Enter class (1-12): ")
        mobile = input("Enter guardian's 10-digit mobile number: ")
        if not self.validate_mobile(mobile):
            print("Invalid mobile number! Must be 10 digits.")
            return

        student_id = self.next_id
        self.next_id += 1


        self.students[student_id] = {"name": name,"age": age,"class": student_class,"guardian_mobile": mobile}
        print(f"Admission successful! Student ID: {student_id}")


    def view_student(self):
        student_id = int(input("Enter student ID: "))
        if student_id in self.students:
            student = self.students[student_id]
            print(f"Student ID: {student_id}")
            print(f"Name: {student['name']}")
            print(f"Age: {student['age']}")
            print(f"Class: {student['class']}")
            print(f"Guardian Mobile: {student['guardian_mobile']}")
        else:
            print("Student not found!")


    def update_student(self):
        student_id = int(input("Enter student ID to update: "))
        if student_id not in self.students:
            print("Student not found!")
            return

        print("1. Update Mobile Number")
        print("2. Update Class")
        choice = input("Enter choice: ")

        if choice == "1":
            mobile = input("Enter new 10-digit mobile number: ")
            if self.validate_mobile(mobile):
                self.students[student_id]['guardian_mobile'] = mobile
                print("Mobile number updated successfully!")
            else:
                print("Invalid mobile number!")
        elif choice == "2":
            student_class = input("Enter new class (1-12): ")
            self.students[student_id]['class'] = student_class
            print("Class updated successfully!")
        else:
            print("Invalid choice!")


    def remove_student(self):
        student_id = int(input("Enter student ID to remove: "))
        if student_id in self.students:
            del self.students[student_id]
            print("Student record removed successfully!")
        else:
            print("Student not found!")
```

```python
school = SchoolManagement()
```

```python
school.new_admission()
```

```python
school.view_student()
```
