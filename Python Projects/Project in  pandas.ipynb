# Assessment2: Retail Sales Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('/content/Retail Data1.xlsx')
```

```python
df1.head(2)
```

```python
'''1. View the structure of the dataset (columns, types, missing values).'''
print(df1.columns)
print(df1.dtypes)
print(df1.isnull().sum())
#
```

```python
'''2. What is the shape (rows, columns) of the dataset? '''
df1.shape
```

```python
'''3. Are there any duplicate records?\ '''
print(df1.duplicated().any())
```

```python
'''4.  Are there any missing or corrupted entries in Ship Date, Order Date, or numeric columns? '''
missing_values = df1.isnull().sum()
print(missing_values)
#
```

```python
'''5.  Convert Order Date and Ship Date to datetime. '''

df1['Order Date'] = pd.to_datetime(df1['Order Date'])
df1['Ship Date'] = pd.to_datetime(df1['Ship Date'])
df1.head(2)
```

```python
'''6.  Check for future or inconsistent shipping dates. '''
future_shipping_dates = df1[df1['Ship Date'] > pd.to_datetime('today')]
print(future_shipping_dates)
```

```python
'''7.  Convert price columns to numeric (remove $ and commas). '''

price_columns = ['Cost Price', 'Retail Price','Profit Margin',
                 'Sub Total','Discount $','Order Total','Shipping Cost','Total']
for i in price_columns:
  df1[i] = df1[i].str.replace('$', '', regex=False)
  pass

df1
```

```python
'''8. What are the unique values in Customer Type and Order Priority? '''
unique_customer_types = df1['Customer Type'].unique()
unique_order_priorities = df1['Order Priority'].unique()
print(unique_customer_types)
print(unique_order_priorities)
#
```

```python
'''9. What are the most common shipping modes? '''
shipping_mode_counts = df1['Ship Mode'].value_counts()
print(shipping_mode_counts)
#
```

```python
'''10. Which cities have the highest number of orders? '''
city_order_counts = df1['City'].value_counts()
print(city_order_counts)
# Sydney has the highest number of orders
```

```python
'''11. What’s the range of order quantities and prices? '''
df1['Order Quantity'].min(), df1['Order Quantity'].max()
price_cols = ['Cost Price', 'Retail Price', 'Sub Total', 'Discount $',
              'Order Total', 'Shipping Cost', 'Total']

df1[price_cols].describe()

```

```python
'''12. Create a new column for shipping duration. '''
df1['Shipping Duration'] = (df1['Ship Date'] - df1['Order Date']).dt.days
df1.head(2)
```

```python
'''13. Are there any orders with zero or negative total or quantity?'''
df1['Order Quantity']=pd.to_numeric(df1['Order Quantity'], errors='coerce')
df1['Total']=pd.to_numeric(df1['Total'], errors='coerce')
zero_total_quantity = df1[(df1['Total'] <= 0) | (df1['Order Quantity'] <= 0)]
print(zero_total_quantity)
```

```python
'''Are all discount percentages matching discount dollar amounts?'''
df1['Discount %']=pd.to_numeric(df1['Discount %'], errors='coerce')
df1['Sub Total']=pd.to_numeric(df1['Sub Total'], errors='coerce')

discount_matching = (df1['Discount %'] * df1['Sub Total']).round(2) == df1['Discount $']
print(discount_matching.all())
```

```python
'''15. Check for mismatches in total calculation’'''
df1['Cost Price'] = pd.to_numeric(df1['Cost Price'], errors='coerce')
df1['Expected Total'] = df1['Order Quantity'] * df1['Cost Price']
mismatches = df1[df1['Total'] != df1['Expected Total']]
print(mismatches)
```

```python
'''16.  Identify top 5 products by order quantity. '''

top_5_products = df1.groupby('Product Name')['Order Quantity'].sum().nlargest(5)
print(top_5_products)
```

```python
'''17. Which Account Manager handled the most revenue? '''
df1['Revenue'] = df1['Order Quantity'] * df1['Cost Price']
manager_revenue = df1.groupby('Account Manager')['Revenue'].sum()
most_revenue_manager = manager_revenue.idxmax()
print(most_revenue_manager)
```

```python
'''18.  What is the average shipping cost by mode? '''
df1['Shipping Cost'] = pd.to_numeric(df1['Shipping Cost'], errors='coerce')
shipping_cost_by_mode = df1.groupby('Ship Mode')['Shipping Cost'].mean()
print(shipping_cost_by_mode)
```

```python
'''19. Find the most profitable product. '''
df1['Profit Margin'] = pd.to_numeric(df1['Profit Margin'], errors='coerce')
most_prfitable_product = df1.groupby('Product Name')['Profit Margin'].mean().idxmax()
print(most_prfitable_product)
```
