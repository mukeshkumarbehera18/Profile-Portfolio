```python
class ClinicAppointment:
    def __init__(self):
        self.time_slots = ["10am", "11am", "12pm", "2pm", "3pm"]
        self.appointments = {}
        self.slot_count = {}

    def book_appointment(self):
        name = input("Enter patient name: ")
        age = input("Enter patient age: ")
        mobile = input("Enter mobile number: ")

        if mobile in self.appointments:
            print(" You already have an appointment booked.")
            return

        doctor = input("Enter preferred doctor: ")

        print("\nAvailable Time Slots:")
        for i, slot in enumerate(self.time_slots, 1):
            print(f"{i}. {slot}")

        slot_choice = int(input("Choose slot (1-5): "))
        if not (1 <= slot_choice <= 5):
            print("Invalid slot choice.")
            return

        chosen_slot = self.time_slots[slot_choice - 1]
        if self.slot_count.get((doctor, chosen_slot), 0) >= 3:
            print(f"Slot {chosen_slot} for Dr.{doctor} is full. Please choose another.")
            return
        self.appointments[mobile] = {"name": name,"age": age,"doctor": doctor,"slot": chosen_slot}
        self.slot_count[(doctor, chosen_slot)] = self.slot_count.get((doctor, chosen_slot), 0) + 1

        print(f"Appointment booked successfully for {name} with Dr.{doctor} at {chosen_slot}.")

    def view_appointment(self):
        mobile = input("Enter your mobile number: ")
        if mobile in self.appointments:
            appt = self.appointments[mobile]
            print("\nAppointment Details:")
            print(f"Name: {appt['name']}")
            print(f"Age: {appt['age']}")
            print(f"Doctor: {appt['doctor']}")
            print(f"Time Slot: {appt['slot']}")
        else:
            print("No appointment found with this mobile number.")

    def cancel_appointment(self):
        mobile = input("Enter your mobile number to cancel: ")
        if mobile in self.appointments:
            appt = self.appointments.pop(mobile)
            doctor, slot = appt['doctor'], appt['slot']
            self.slot_count[(doctor, slot)] -= 1
            print(f"Appointment with Dr.{doctor} at {slot} has been cancelled.")
        else:
            print("No appointment found with this mobile number.")

    def doctor_availability(self):
        doctor = input("Enter doctor's name to check availability: ")
        print(f"\nAvailability for Dr.{doctor}:")
        for slot in self.time_slots:
            count = self.slot_count.get((doctor, slot), 0)
            available = 3 - count
            print(f"{slot}: {available} slots available")
```

```python
clinic = ClinicAppointment()
```

```python
clinic.book_appointment()
```

```python
clinic.view_appointment()
```

```python
clinic.doctor_availability()
```

```python
clinic.cancel_appointment()
```
