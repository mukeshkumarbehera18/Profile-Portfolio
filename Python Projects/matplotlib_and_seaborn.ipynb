```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
df1 = df.groupby('cut')['price'].sum().reset_index()

#plt.bar(df1['cut'],df1['price'])

#plt.figure(figsize=(10,5))
plt.plot(df1['cut'],df1['price'],marker="*",color='red')#linestyle='--')
plt.title('Cut wise price')
plt.xlabel("cut")
plt.grid(True)
plt.ylabel("Total Price")
plt.show()
```

```python
sns.lineplot(data=df1, x=df['cut'], y=df['price'], marker='o', color='green')
plt.title("cut wise price")
plt.show()
```

```python
plt.pie(df1['price'], labels=df1['cut'], autopct='%1.1f%%')
plt.show()
```

```python
cut_count = df['cut'].value_counts()
plt.pie(cut_count, labels=cut_count.index)#, autopct='%1.1f%')
plt.show()
```

```python
plt.hist(df['table'], bins=60, color='skyblue', edgecolor='black' )#bins=50)
plt.show()
```

```python
sns.pairplot(df, vars=['carat', 'depth', 'table', 'price'], palette='Set2', hue='cut')
```
